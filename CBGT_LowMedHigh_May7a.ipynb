```python
# bdf_filename = os.path.join(savedir, 'cbgt_behavdf_all_{}.csv'.format(parentDir))
# bdf_filename2 = os.path.join(savedir, 'cbgt_behavdf_all_regs_{}.csv'.format(parentDir))
# bdf = pd.read_csv(bdf_filename)

dL = dlx.reset_index(drop=True)
iL = ilx.reset_index(drop=True)
dR = drx.reset_index(drop=True)
iR = irx.reset_index(drop=True)
bdf['dL_rate_rt'] = np.nan; bdf['iL_rate_rt'] = np.nan
bdf['dR_rate_rt'] = np.nan; bdf['iR_rate_rt'] = np.nan 

for c in conds:
    
    trials = bdf[(bdf.cond==c)&(~bdf.rt.isna())].trial.values - 1
    rt = bdf[(bdf.cond==c)&(~bdf.rt.isna())]['rtms'].astype(int).reset_index(drop=True).values - 200
    acc = bdf[(bdf.cond==c)&(~bdf.rt.isna())]['acc'].astype(int).reset_index(drop=True).values
    
    leftD1 = np.asarray([dL[(dL.cond==c)&(dL.trial==t)].ix[:, 200:].values[:, rt[i]] for i, t in enumerate(trials)])
    leftD2 = np.asarray([iL[(iL.cond==c)&(iL.trial==t)].ix[:, 200:].values[:, rt[i]] for i, t in enumerate(trials)])
    rightD1 = np.asarray([dR[(dR.cond==c)&(dR.trial==t)].ix[:, 200:].values[:, rt[i]] for i, t in enumerate(trials)])
    rightD2 = np.asarray([iR[(iR.cond==c)&(iR.trial==t)].ix[:, 200:].values[:, rt[i]] for i, t in enumerate(trials)])
    
    DLFull = [dL[(dL.cond==c)&(dL.trial==t)].ix[:, 200:].dropna(axis=1).values for i, t in enumerate(trials)]
    ILFull = [iL[(iL.cond==c)&(iL.trial==t)].ix[:, 200:].dropna(axis=1).values for i, t in enumerate(trials)]
    DRFull = [dR[(dR.cond==c)&(dR.trial==t)].ix[:, 200:].dropna(axis=1).values for i, t in enumerate(trials)]
    IRFull = [iR[(iR.cond==c)&(iR.trial==t)].ix[:, 200:].dropna(axis=1).values for i, t in enumerate(trials)]

    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'dL_rate_rt'] = leftD1
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'iL_rate_rt'] = leftD2
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'dR_rate_rt'] = rightD1
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'iR_rate_rt'] = rightD2
    
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'D_LR_diff'] = leftD1 - rightD1
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'I_LR_mean'] = np.column_stack([leftD2, rightD2]).mean(axis=1)
    
    d_diff_list, di_diff_list, id_diff_list, imax_list = [], [], [], []
    d_diff_sum_list = []; d_lrdiff_sum_list = []; di_ldiff_sum_list = []
    di_lrdiff_sum_list = []; di_rdiff_sum_list = []; di_lrdiff_sum_list = []
    
    for i, t in enumerate(trials):  
        d_lrdiff_sum_list.append(np.sum(DLFull[i] - DRFull[i]))
        di_ldiff_sum_list.append(np.sum(DLFull[i] - ILFull[i]))
        di_rdiff_sum_list.append(np.sum(DRFull[i] - IRFull[i]))
        di_lrdiff_sum_list.append(np.sum(DLFull[i] - ILFull[i]) - np.sum(DRFull[i] - IRFull[i]))
        
        if acc[i]:
            d_diff_list.append(leftD1[i] - rightD1[i])
            id_diff_list.append(leftD2[i] - leftD1[i])
            di_diff_list.append(np.sum(DLFull[i] - ILFull[i]))
            d_diff_sum_list.append(np.sum(DLFull[i] - DRFull[i]))
            imax_list.append(leftD2[i])
            
        else:
            d_diff_list.append(rightD1[i] - leftD1[i])
            d_diff_sum_list.append(np.sum(DRFull[i] - DLFull[i]))
            id_diff_list.append(rightD2[i] - rightD1[i])
            di_diff_list.append(np.sum(DRFull[i] - IRFull[i]))
            imax_list.append(rightD2[i])
    
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'D_lrdiff_sum'] = d_lrdiff_sum_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'DI_ldiff_sum'] = di_ldiff_sum_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'DI_rdiff_sum'] = di_rdiff_sum_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'DI_lrdiff_sum'] = di_lrdiff_sum_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'ID_diff_chosen'] = id_diff_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'DI_diff_chosen'] = di_diff_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'D_diff'] = d_diff_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'D_diff_sum'] = d_diff_sum_list
    bdf.ix[(bdf.cond==c)&(~bdf.rt.isna()), 'I_max_chosen'] = imax_list
```
